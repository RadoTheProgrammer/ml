```python
import pandas as pd
myDataset=pd.read_csv("Data/ecoles-creches-idf.csv")
myDataset
```

```python
newDataset=myDataset.melt(
    ["ID","geometry"],
    ["PM10_2012","PM10_2013","PM10_2014","PM10_2015","PM10_2016","PM10_2017"],
    "PM10-Année", "Valeur")
newDataset
```

```python
myDataset=myDataset[["ID","geometry"]]
myDataset
```

```python
import re

def get_latitude(s, flag):
    res = re.split(r"\(|\,|\)",s)
    if flag == 0:
        res = res[1]
    else:
        res = res[2]
    return float(res)

myDataset["Latitude"] = myDataset["geometry"].apply(get_latitude, args=(0,))
myDataset
```
