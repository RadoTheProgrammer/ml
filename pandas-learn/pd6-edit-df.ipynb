```python
import pandas as pd

myDataset = pd.read_csv('Data/ozone.csv')
```

```python
myDataset["Resultat"] = False
myDataset
```

```python
myDataset["Pow2_O3obs"] = myDataset["O3obs"]**2
myDataset
```

```python
myDataset["Total"] = myDataset.iloc[:,5:7].sum(axis=1) #axis=1 pour sommer les colonnes par ligne
myDataset
```

```python
myDataset["Total"] = myDataset.loc[:,["NO2","NO"]].sum(axis=1)
myDataset
```

```python
columns=list(myDataset.columns)
newDataset = myDataset[columns[0:7]+[columns[-1]]+columns[7:-1]]
newDataset
```

```python
myDataset=myDataset.drop(columns=["Pow2_O3obs","Total"])
myDataset
```

```python
myDataset["Resultat"] = False
columns = list(myDataset.columns)
myDataset = myDataset[columns[:3]+[columns[-1]]+columns[3:-1]]
dif=abs(myDataset["O3obs"]-myDataset["MOCAGE"])
myDataset.loc[dif < 2.5, "Resultat"] = True
myDataset
```

```python
myDataset = myDataset.drop(columns=["Resultat"])
```

```python
myDataset.tail(3)
```

```python
new_row = {
    "JOUR":0, 
    "O3obs":10,
    "MOCAGE":10,
    "TEMPE":10,
    "RMH20":10,
    "NO2":10,
    "NO":10,
    "STATION":"Aix",
    "VentMOD":10,
    "VentANG":10
}
myDataset = pd.concat([myDataset,pd.DataFrame([new_row])])

myDataset.tail(3)
```
